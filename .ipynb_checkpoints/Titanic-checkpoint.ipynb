```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
titanic_data = pd.read_csv('Titanic.csv')

# Handle missing values
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

# Drop columns that are irrelevant or have too many missing values
titanic_data.drop(columns=['Cabin', 'PassengerId', 'Name', 'Ticket'], inplace=True)

# One-hot encode categorical variables
titanic_data = pd.get_dummies(titanic_data, columns=['Sex', 'Embarked'], drop_first=True)

# Scale numerical features
scaler = StandardScaler()
titanic_data[['Age', 'Fare']] = scaler.fit_transform(titanic_data[['Age', 'Fare']])

# Save the preprocessed dataset (optional)
titanic_data.to_csv('Preprocessed_Titanic.csv', index=False)

# Display the first few rows of the preprocessed data
print(titanic_data.head())
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the preprocessed dataset
titanic_data = pd.read_csv('Preprocessed_Titanic.csv')

# Check for missing values
if titanic_data.isnull().sum().sum() > 0:
    # Fill any remaining NaN values in numerical columns with the median
    for col in titanic_data.select_dtypes(include=['float64', 'int64']).columns:
        titanic_data[col].fillna(titanic_data[col].median(), inplace=True)
    
    # Fill remaining NaNs in categorical columns with the mode
    for col in titanic_data.select_dtypes(include=['object']).columns:
        titanic_data[col].fillna(titanic_data[col].mode()[0], inplace=True)

# Extract features for clustering (excluding 'Survived' if present)
features = titanic_data.drop(columns=['Survived'], errors='ignore')

# Scale numerical features
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Elbow Method to find the optimal number of clusters
inertia = []
cluster_range = range(1, 11)
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()

# Perform K-Means clustering with the chosen number of clusters (e.g., 3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
titanic_data['Cluster'] = kmeans.fit_predict(features)

# Display the first few rows with cluster labels
print(titanic_data.head())

# Save the dataset with cluster labels (optional)
titanic_data.to_csv('Titanic_with_Clusters.csv', index=False)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Perform PCA to reduce dimensions to 2D
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(features)

# Add the reduced dimensions to the dataframe for visualization
titanic_data['PCA1'] = reduced_features[:, 0]
titanic_data['PCA2'] = reduced_features[:, 1]

# Plot the clusters
plt.figure(figsize=(10, 6))
for cluster_id in titanic_data['Cluster'].unique():
    cluster_data = titanic_data[titanic_data['Cluster'] == cluster_id]
    plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], label=f'Cluster {cluster_id}', alpha=0.6)

# Add plot decorations
plt.title('K-Means Clusters Visualization (2D PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
